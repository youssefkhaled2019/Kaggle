```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
import  warnings; warnings.filterwarnings('ignore')
%matplotlib inline 


```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data=pd.read_csv('../input/digit-recognizer/train.csv')
test_data=pd.read_csv('../input/digit-recognizer/test.csv')
```

```python
train_data.head()
```

```python
train_data['label'].value_counts().sort_index()
```

```python
X=train_data.drop(['label'],axis=1)
Y=train_data['label']
```

```python
import math

fig,ax =plt.subplots(5,5,figsize=(10,10))
for i in range(1,26):
  img=X.iloc[i].values
  img=img.reshape((28,28))
  m=math.ceil(i/5)-1
  n=[0,1,2,3,4]*5
  ax[m,n[i-1]].imshow(img)
```

```python
plt.hist(X.iloc[0].values)
plt.show()
```

```python
X/=255
test_data/=255
```

```python
pca=PCA(n_components=200)
pca.fit(X)
plt.plot(pca.explained_variance_ratio_)
plt.title('explained_variance')
```

```python
pca=PCA(n_components=150)
new_traing_data=pd.DataFrame(pca.fit_transform(X))
new_test_data=pd.DataFrame(pca.transform(test_data))
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_traing_data, Y, test_size=0.20,stratify=train_data['label'], random_state=42, shuffle =True)
```

```python
svc_mod=SVC()
svc_mod.fit(x_train,y_train)
pred=svc_mod.predict(x_test)
# # print("trsing score",svc_model.score(x_train,y_train))#trsing score 0.9893452380952381
print("test score ",svc_mod.score(x_test,y_test))#test score  0.9841666666666666
print(confusion_matrix(pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
svc_model_1=SVC()


pramter={"kernel":["linear","poly"],"gamma":['auto',"scale"]}
GridSearch=GridSearchCV(estimator=svc_model_1,param_grid=pramter,cv=3,return_train_score=True)
GridSearch.fit(new_traing_data,Y)
print('Best Score is :', GridSearch.best_score_)
print('Best Parameters are :', GridSearch.best_params_)

# Best Score is : 0.9779285714285714
# Best Parameters are : {'gamma': 'scale', 'kernel': 'poly'}
```

```python
svc_model=SVC()
svc_model.fit(new_traing_data,Y)
svc_model.score(new_traing_data,Y)
```

```python
data_pred=svc_model.predict(new_test_data)
```

```python
result={'ImageId':test_data.index+1,'Label':data_pred}
```
